<div align="left"><span style="font-family:Arial Black;font-size:12px;color:black">Données manquantes : </span></div>

```python
def graph_missing_data(df:pd.DataFrame) :
    plt.figure(figsize=(16, 4))
    (df.count() - df.shape[1]).plot.bar()
    plt.show()
```
